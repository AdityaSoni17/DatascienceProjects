```python
import pandas as pd
import numpy as np
```

```python
movies_df = pd.read_csv('/cxldata/skin_disease_1/sudeep/recommended_system/archive/tmdb_5000_movies.csv')
```

```python
movies_df.head()
```

```python
credit_df = pd.read_csv('/cxldata/skin_disease_1/sudeep/recommended_system/archive/tmdb_5000_credits.csv')
```

```python
credit_df.head()
```

```python
credit_df_rename = credit_df.rename(columns={'movie_id':'id'})
```

```python
credit_df_rename.head()
```

```python
movies_merged = movies_df.merge(credit_df_rename, on ='id')
```

```python
movies_merged
```

```python
movies_merged = movies_merged.drop(columns = ['title_x','title_y','homepage','status'])
```

```python
movies_merged.head()
```

```python
####### weighted average = (R*v + C*m)/m + v

# v = number of votes count(i.e. people have voted)
# R = average number of votes
# m = minimum number of votes count participate
# C = should be above average of average rating 
```

```python
v = movies_merged['vote_count']
R = movies_merged['vote_average']
m = movies_merged['vote_count'].quantile(0.70)
C = movies_merged['vote_average'].mean()
```

```python
movies_merged['weighted_average'] =  ((R*v) + (C*m))/(m +v)
```

```python
movies_merged.head()
```

```python
movies_merged_sort = movies_merged.sort_values('weighted_average', ascending=False)
```

```python
movies_merged_sort.head()
```

```python
movies_merged_final = movies_merged_sort[['original_title','vote_average','vote_count','weighted_average']]
```

```python
movies_merged_final
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
movie_merged_scale = scaler.fit_transform(movies_merged[['popularity','weighted_average']])
```

```python
movie_normalized_df=pd.DataFrame(movie_merged_scale,columns=['weighted_average','popularity'])
```

```python
movie_normalized_df.head()
```

```python
movies_merged[['normalized_weighted_average','normalized_popularity']] = movie_normalized_df
```

```python
movies_merged
```

```python
movies_merged['score'] = movies_merged['normalized_weighted_average'] * 0.5 + movies_merged['normalized_popularity']*0.5
```

```python
movies_merged_final = movies_merged.sort_values('score',ascending=False)
```

```python
movies_merged_final.head()
```
